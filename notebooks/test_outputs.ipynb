# Adding an extra key

```python
import torch

def add_key_to_checkpoint(checkpoint_path, key_name, key_value):
    # Load the checkpoint
    checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))

    # Add the new key
    checkpoint[key_name] = key_value

    # Save the modified checkpoint back to the same file
    torch.save(checkpoint, checkpoint_path)
    print(f"Added key '{key_name}' to checkpoint: {checkpoint_path}")

# Specify the checkpoint path
import os

# Directory containing the checkpoints
checkpoints_dir = "/mnt/51eb0667-f71d-4fe0-a83e-beaff24c04fb/om/DiT/results/003-DiT-XL-2/checkpoints"

# Iterate over all checkpoint files in the directory
for checkpoint_filename in os.listdir(checkpoints_dir):
    checkpoint_path = os.path.join(checkpoints_dir, checkpoint_filename)
    
    # Add the key 'is_depth' with value True
    add_key_to_checkpoint(checkpoint_path, "is_depth", True)
```

```python
# Remove opt from ckpt as well 
import torch

def modify_checkpoint(checkpoint_path, key_name, key_value, save_compressed=True):
    # Load the checkpoint
    checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))

    # Remove the optimizer state if present
    if "opt" in checkpoint:
        del checkpoint["opt"]

    # Add the new key
    checkpoint[key_name] = key_value

    # Save the modified checkpoint with compression if specified
    torch.save(checkpoint, checkpoint_path, _use_new_zipfile_serialization=save_compressed)
    print(f"Modified checkpoint saved to: {checkpoint_path}")

# Specify the checkpoint path
checkpoints_dir = "/mnt/51eb0667-f71d-4fe0-a83e-beaff24c04fb/om/DiT/results/1-epoch-not-valid-mask/checkpoints"

# Iterate over all checkpoint files in the directory
for checkpoint_filename in os.listdir(checkpoints_dir):
    checkpoint_path = os.path.join(checkpoints_dir, checkpoint_filename)
    
    # Add the key 'is_depth' with value True
    add_key_to_checkpoint(checkpoint_path, "is_depth", True)


checkpoints_dir = "/mnt/51eb0667-f71d-4fe0-a83e-beaff24c04fb/om/DiT/results/1-epoch-valid-mask/checkpoints"

# Iterate over all checkpoint files in the directory
for checkpoint_filename in os.listdir(checkpoints_dir):
    checkpoint_path = os.path.join(checkpoints_dir, checkpoint_filename)
    
    # Add the key 'is_depth' with value True
    add_key_to_checkpoint(checkpoint_path, "is_depth", True)

```
